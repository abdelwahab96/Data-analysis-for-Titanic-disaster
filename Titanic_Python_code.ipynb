### Cabin attribute has many missing values .. it's better to drop this column

```python
train = train.drop(columns=['Cabin'])
train
```

* from previous analysis work for other analysts .. they recomend to get the name title of each person to check if the standard living affects on the survived percentage or doesn't 

```python
t_name = train['Name'].apply(lambda x: x.split('.')[0]) 
train['Name Title'] = t_name.apply(lambda x: x.split(',')[1])
train
```

```python
tita = train[train['Survived'] == 1]

tita
```

```python
# get the number of survived people grouped by their name titles (survived=1)
tit = tita[['Name Title', 'Survived']].groupby(["Name Title"], as_index=False).count()
tit
```

```python
# dataframe for dead people
titb = train[train['Survived'] == 0]
titb
```

```python
# get the number of dead people grouped by their name titles (survived=0)
tit2 = titb[['Name Title', 'Survived']].groupby(["Name Title"], as_index=False).count()
tit2
```

```python
#it's clear nwo that ['PassengerId','Name','Ticket'] are not important to this analysis. thus, it's better to drop them
train = train.drop(columns=['PassengerId','Name','Ticket'])
train
```

# visualising heatmap correlation 

```python
#making mask for upper side
df_corr = train.corr()
mask = np.triu(df_corr)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(data=df_corr, linewidths=1,annot=True,fmt='0.1f',mask=mask)
```

```python
#changing Pclass attributes to ['First','Middle','Lowest']
for ind, k in enumerate(train['Pclass']):
    if k == 1: 
        train['Pclass'][ind] = 'First'
    elif k == 2:
        train['Pclass'][ind] = 'Middle'
    elif k ==3:
        train['Pclass'][ind] = 'Lowest'

train
```
